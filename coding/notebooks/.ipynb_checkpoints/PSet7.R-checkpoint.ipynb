#### (c)

If we estimate the following model:

\begin{align}
Y = X_0 \beta + Z \gamma + \epsilon \\
Z = X_0 \beta^\star + Y \gamma^\star + \epsilon^\star
\end{align}

We have two potential vectors of independent variables:
\begin{align}
X = \begin{bmatrix}
X_0 & Z
\end{bmatrix} \\
X^\star = \begin{bmatrix}
X_0 & Y
\end{bmatrix}
\end{align}

We denote $M_X$ as the annihilator matrix for the matrix $X_0$. We denote $M$ as the annihilator matrix for $X$ and $M^\star$ as the annihilator matrix for $X^\star$

We are interested in comparing the t-statistic for Z in the first equation with the t-statistic for $Y$ in the second. From Frisch-Waugh-Lowell (FWL), we can get the estimated values for $\gamma$ and $\gamma^\star$:
\begin{align}
\hat \gamma = (Z' M_X Z)^{-1}Z'M_X Y \\
\hat{\gamma^\star} = (Y' M_X Y)^{-1}Y'M_X Z
\end{align}

If we define the errors for both regressions as $e$ and $e^\star$, those will be:
\begin{align}
e = M Y \\
e^\star = M^\star Z
\end{align}

But we can use FWL for the regression to find that $e$ is equivalent to the residuals of the regression of $M_X Z$ on $M_X Y$ and the converse for $e^\star$:
\begin{align}
e = M_X Y - M_X Z (Z' M_X Z)^{-1}Z'M_X Y \\
e^\star = M_X Z - M_X Y (Y' M_X Y)^{-1}Y'M_X Z
\end{align}

We can estimate the variance of the errors under the homoscedasticity assumption. For the first equation we have the following:
\begin{align}
s^2 &=& \frac{e'e}{(n-k)} & & \\
&=& \frac{Y'M_X(I - M_XZ(Z'M_XZ)^{-1}Z'M_X)M_XY}{(n-k)} & & \text{By definition of $e$} \\
&=& \frac{Y'M_X(I \times (Z'M_XZ) - M_XZZ'M_X)M_XY}{(Z'M_XZ)(n-k)} & & \text{As $(Z'M_XZ)$ is a scalar} \\
&=& \frac{(Y'M_XY) \times (Z'M_XZ) - Y'M_XZZ'M_XY}{(Z'M_XZ)(n-k)} & & \text{Expanding the expression} \\
&=& \frac{(Y'M_XY) \times (Z'M_XZ) - (Z'M_XY)^2}{(Z'M_XZ)(n-k)} & & \text{As $Z'M_XY$ is a scalar} \\
\end{align}
The case of the estimate of the variance of the error in the second equation is similar:
\begin{align}
s^2_\star&=& \frac{(Y'M_XY) \times (Z'M_XZ) - (Z'M_XY)^2}{(Y'M_XY)(n-k)} & &  \\
\end{align}


We can use these estimates to get a value for the variances of $\gamma$ and $\gamma^\star$. Again, we will use FWL to get the variances using the residuals from the regressions on $X_0$ using the annihilator matrix $M_X$:

\begin{align}
Var(\hat \gamma) = s^2 (Z'M_XZ)^{-1} = \frac{(Y'M_XY) \times (Z'M_XZ) - (Z'M_XY)^2}{(Z'M_XZ)^2(n-k)} \\
Var(\hat \gamma^\star) = s^2_\star (Y'M_XY)^{-1} = \frac{(Y'M_XY) \times (Z'M_XZ) - (Z'M_XY)^2}{(Y'M_XY)^2(n-k)}
\end{align}

Now, we can calculate the t-statistic for $\hat \gamma$ and $\hat \gamma^\star$:
\begin{align}
t &=& \frac{\hat \gamma}{s} \\
&=& \frac{(Z' M_X Z)^{-1}Z'M_X Y}{(Z'M_XZ)^{-1}\sqrt{((Y'M_XY) \times (Z'M_XZ) - (Z'M_XY)^2)/(n-k)}} \\
&=& \frac{Z'M_X Y}{\sqrt{((Y'M_XY) \times (Z'M_XZ) - (Z'M_XY)^2)/(n-k)}}
\end{align}

\begin{align}
t^\star &=& \frac{\hat \gamma^\star}{s_\star} \\
&=& \frac{(Y' M_X Y)^{-1}Y'M_X Z}{(Y'M_XY)^{-1}\sqrt{((Y'M_XY) \times (Z'M_XZ) - (Z'M_XY)^2)/(n-k)}} \\
&=& \frac{Y'M_X Z}{\sqrt{((Y'M_XY) \times (Z'M_XZ) - (Z'M_XY)^2)/(n-k)}}
\end{align}

Because $Y'M_X Z$ is a scalar $Y'M_X Z = Z'M_X Y$, which is the final step to prove that $t = t^\star$.

### Question 2

We have the following set-up as the true model
:

$$y_i = \beta_1 x_i + \beta_2 w_i + \epsilon_i$$

The variables are demeaned by the population mean, so we can replace the covariance between them as the expectation of the product. Along this exercise, I'll be using the following equivalences:
$$Var(m) = E(m^2)$$
$$Cov(m,n) = E(m)E(n)$$
In the case of demeaning using the sample mean, the results are similar. Still, additional non-informative steps are needed to prove the results, like proving that the sample mean converges in probability to the true mean or that the sample covariance converges in probability to the true covariance.

To solve this question, I need to define elements. The short estimator will be denoted as $\hat\beta_s$ and the long estimator will be denoted as $\hat \beta_l$. The short estimator comes from the regression of $y_i$ in $x_i$:

$$\hat\beta_s = \frac{\sum_{i=1}^n x_iy_i}{\sum_{i=1}^n x_i^2}= \frac{\sum_{i=1}^n x_iy_i/n}{\sum_{i=1}^n x_i^2/n}$$

The long regression comes from the OLS formula. I define the vector $Z_i' = \begin{bmatrix}
y_i & w_i
\end{bmatrix}$. The long regression is the result of the following calculation:

$$\hat\beta = (\sum_{i=1}^n Z_i Z_i')^{-1} \sum_{i=1}^n Z_i y_i = (\sum_{i=1}^n Z_i Z_i'/n)^{-1} \sum_{i=1}^n Z_i y_i/n$$

The first element from this vector $\hat \beta$ is the long regression estimation $\hat \beta_l$.

Using the WLLN and CMT, it is easy to show that the short regression estimate converges in probability to:

\begin{align*}
\hat\beta_s &=& \frac{\sum_{i=1}^n x_iy_i/n}{\sum_{i=1}^n x_i^2/n} \\
\hat\beta_s &\xrightarrow[p]{}& \frac{E(x_iy_i)}{E(x_i^2)} \\
\end{align*}

To get the asymptotic distribution of this estimator, we can use the true specification. If we want to apply CLT, we will need additional conditions on the variables, like the relationship between $x_i$ and $\epsilon_i$.

Using the WLLN and CMT, it is easy to show that the long regression estimate converges to:

\begin{align*}
\hat\beta &=& (\sum_{i=1}^n Z_i Z_i'/n)^{-1} \sum_{i=1}^n Z_i y_i/n \\
 \hat\beta &\xrightarrow[p]{}& E(Z_iZ_i') ^{-1}E(Z_i y_i) \\
 \hat\beta &\xrightarrow[p]{}& \frac{1}{E(x_i^2)E(w_i^2) - E(x_iw_i)^2}
 \begin{bmatrix}
E(w_i^2) & -E(x_i w_i) \\
-E(x_i w_i) & E(x_i^2)
\end{bmatrix} 
\begin{bmatrix}
E(x_i y_i) \\
E(w_i y_i)
\end{bmatrix} 
\end{align*}

The first element of this vector is $\hat \beta_l$. From the matrix computation, it is easy to show that:

\begin{align*}
\hat\beta_l &\xrightarrow[p]{}& \frac{E(w_i^2)E(x_i y_i) - E(x_i w_i)E(w_i y_i)}{E(x_i^2)E(w_i^2) - E(x_iw_i)^2}
\end{align*}

### 2.1

In this setup:
\begin{align*}
\epsilon_i \perp x_i \to E(\epsilon_i \ x_i) = 0 \\
\epsilon_i \perp w_i \to E(\epsilon_i \ w_i) = 0 \\
E(x_i w_i) \ne 0
\end{align*}

We can make the computation of the cross expectations using these assumptions:
\begin{align*}
E(x_iy_i) &=& E(x_i (\beta_1 x_i + \beta_2 w_i + \epsilon_i)) \\
 &=& \beta_1E(x_i^2) + \beta_2 E(x_i w_i) \\
E(w_iy_i) &=& E(w_i (\beta_1 x_i + \beta_2 w_i + \epsilon_i)) \\
 &=& \beta_1E(x_i w_i) + \beta_2 E(w_i^2) \\
\end{align*}


##### (a)

Using the previous results, it is easy to show that:

\begin{align*}
\hat\beta_s &\xrightarrow[p]{}& \frac{E(x_iy_i)}{E(x_i^2)} \\
\hat\beta_s &\xrightarrow[p]{}& \frac{\beta_1E(x_i^2) + \beta_2 E(x_i w_i)}{E(x_i^2)} \\
\hat\beta_s &\xrightarrow[p]{}& \beta_1 + \beta_2 \frac{E(x_i w_i)}{E(x_i^2)} \\
\end{align*}

Thus, the short regression is inconsistent.

##### (b)

Using the previous results

\begin{align*}
\hat\beta_l &\xrightarrow[p]{}& \frac{E(w_i^2)E(x_i y_i) - E(x_i w_i)E(w_i y_i)}{E(x_i^2)E(w_i^2) - E(x_iw_i)^2} \\
\hat\beta_l &\xrightarrow[p]{}& \frac{E(w_i^2)(\beta_1E(x_i^2) + \beta_2 E(x_i w_i)) - E(x_i w_i)E(\beta_1E(x_i w_i) + \beta_2 E(w_i^2))}{E(x_i^2)E(w_i^2) - E(x_iw_i)^2} \\
\hat\beta_l &\xrightarrow[p]{}& \frac{\beta_1(E(w_i^2)E(x_i^2) - E(x_i w_i)^2)}{E(x_i^2)E(w_i^2) - E(x_iw_i)^2} \\
\hat\beta_l &\xrightarrow[p]{}& \beta_1
\end{align*}
Thus, the long regression is consistent.

##### (c)

It is irrelevant as the short regression estimator is inconsistent.

### 2.2

In this setup:
\begin{align*}
\epsilon_i \perp x_i \to E(\epsilon_i \ x_i) = 0 \\
x_i \perp w_i \to E(x_i \ w_i) = 0 \\
E(\epsilon_i w_i) \ne 0
\end{align*}

We can make the computation of the cross expectations using these assumptions:
\begin{align*}
E(x_iy_i) &=& E(x_i (\beta_1 x_i + \beta_2 w_i + \epsilon_i)) \\
 &=& \beta_1E(x_i^2)\\
E(w_iy_i) &=& E(w_i (\beta_1 x_i + \beta_2 w_i + \epsilon_i)) \\
 &=&  \beta_2 E(w_i^2) + E(w_i \epsilon_i)\\
\end{align*}


##### (a)

Using the previous results, it is easy to show that:

\begin{align*}
\hat\beta_s &\xrightarrow[p]{}& \frac{E(x_iy_i)}{E(x_i^2)} \\
\hat\beta_s &\xrightarrow[p]{}& \frac{\beta_1E(x_i^2)}{E(x_i^2)} \\
\hat\beta_s &\xrightarrow[p]{}& \beta_1 \\
\end{align*}

Thus, the short regression is consistent.

##### (b)

Using the previous results

\begin{align*}
\hat\beta_l &\xrightarrow[p]{}& \frac{E(w_i^2)E(x_i y_i) - E(x_i w_i)E(w_i y_i)}{E(x_i^2)E(w_i^2) - E(x_iw_i)^2} \\
\hat\beta_l &\xrightarrow[p]{}& \frac{E(w_i^2)(\beta_1E(x_i^2)) - E(x_i w_i)E(\beta_2 E(w_i^2) + E(w_i \epsilon_i))}{E(x_i^2)E(w_i^2) - E(x_iw_i)^2} \\
\hat\beta_l &\xrightarrow[p]{}& \frac{\beta_1 E(w_i^2)E(x_i^2)}{E(x_i^2)E(w_i^2)} && \text{As $E(w_i x_i) = 0$}\\
\hat\beta_l &\xrightarrow[p]{}& \beta_1
\end{align*}
Thus, the long regression is consistent.

##### (c)

Both estimators are consistent so we can make the comparison between their variances. We need to apply the CLT to iid processes to obtain the asymptotic distribution. In the case of the short regression, we can re-arrange the terms to obtain a suitable expression to use CLT.

\begin{align*}
\sqrt{n}(\hat\beta_s - \beta_1) &=& \sqrt{n} (\frac{\sum_{i=1}^n x_iy_i/n}{\sum_{i=1}^n x_i^2/n} -\beta_1) & & \\
 &=& \sqrt{n} (\frac{\sum_{i=1}^n x_i(\beta_1 x_i + \beta_2 w_i + \epsilon_i)/n}{\sum_{i=1}^n x_i^2/n} -\beta_1) & & \text{Using the true model} \\
  &=&  \frac{\sqrt{n}(\sum_{i=1}^n x_i(\beta_2 w_i + \epsilon_i))/n}{\sum_{i=1}^n x_i^2/n} & & \text{Simplifying terms} \\
\end{align*}

We know $w_i$, $x_i$, and $w_i$ are demeaned, so they have mean $0$. The expected value for the numerator expression (suitable for CLT) is:
\begin{align*}
E(x_i(\beta_2 w_i + \epsilon_i)) &=& \beta_2 E(x_i w_i) + E(x_i \epsilon_i) &&  \\
&=& 0 && \text{As $x_i$ is independent of $w_i$ and $\epsilon_i$} \\
Var(x_i(\beta_2 w_i + \epsilon_i)) &=& E(x_i^2 (\beta_2 w_i + \epsilon_i)^2) \\
&=& E(x_i^2) E((\beta_2 w_i + \epsilon_i)^2) && \text{As $x_i$ is independent of $w_i$ and $\epsilon_i$}
\end{align*}
 Thus:

\begin{align*}
\sqrt{n}(\sum_{i=1}^n x_i(\beta_2 w_i + \epsilon_i))/n &\xrightarrow[]{d}& N(0, E(x_i^2) E((\beta_2 w_i + \epsilon_i)^2)) & & \text{By CLT}\\
\sum_{i=1}^n x_i^2/n &\xrightarrow[]{p}& E(x_i^2)  & & \text{By WLLN} \\
\sqrt{n}(\hat\beta_s - \beta_1)  &=& \frac{\sqrt{n}(\sum_{i=1}^n x_i(\beta_2 w_i + \epsilon_i))/n}{\sum_{i=1}^n x_i^2/n} \\
\sqrt{n}(\hat\beta_s - \beta_1)  &\xrightarrow[]{d}& N(0, E((\beta_2 w_i + \epsilon_i)^2)/E(x_i^2)) & & \text{By CMT-Slutsky}\\
\end{align*}

In the case of the long regression, to make the calculation easier, I am going to rewrite the true model using a new term $v_i$ that is linear independent of $x_i$ and $w_i$:

\begin{align*}
y_i &=& \beta_1 x_i + \beta_2 w_i + \epsilon_i \\
&=& \beta_1 x_i + (\beta_2 + \beta_3) w_i + v_i \\
\beta_3 &=& E(w_i \epsilon_i)/E(w_i^2) \\
v_i &=& \epsilon_i - \beta_3 w_i \\
E(v_i) &=& 0 \\
E(v_i w_i) &=& 0 \\
E(v_i x_i) &=& 0 \\
E(v_i^2) &=& E(\epsilon_i^2 - 2 \beta_3 w_i \epsilon_i + \beta_3^2 w_i^2) \\
&=& E(\epsilon_i^2)(1 - E(\epsilon_i w_i)^2/(E(\epsilon_i ^2)E(w_i^2)) \\
&=& E(\epsilon_i^2)(1 - \rho_{w_i,\epsilon_i}^2) \\
\end{align*}

$\rho_{w_i,\epsilon_i}$ is the correlation coefficient for $w_i$ and $\epsilon_i$.

If we estimate this new version of the equation, we will have consistent estimators for $\beta_1$ and $\beta_2 + \beta_3$, but we will not be able to have an estimator for $\beta_2$. We can apply the CLT to these estimators. We are not interested in $\beta_2$ nor its estimator. If we denote $\hat \beta$ and $\beta$ the estimator and the true parameters for this regression, we can easily check that:
\begin{align*}
\sqrt{n} (\hat \beta - \beta ) = \sqrt{n} 
(\begin{bmatrix} \hat \beta_1 \\ \hat{\beta_2 + \beta_3} \end{bmatrix} -
\begin{bmatrix} \beta_1 \\ \beta_2 + \beta_3 \end{bmatrix} 
) \xrightarrow{d} N(0, E(Z_iZ_i') ^{-1} E(v_i^2 Z_i Z_i') E(Z_iZ_i') ^{-1})
\end{align*}
We can get the values for the elements of the variance of this expression:
\begin{align*}
E(Z_iZ_i') &=& E(\begin{bmatrix} x_i \\ w_i \end{bmatrix} \begin{bmatrix} x_i & w_i \end{bmatrix}) \\
&=& \begin{bmatrix} E(x_i^2) & E(x_iw_i) \\ E(x_iw_i) & E(w_i^2) \end{bmatrix} \\
&=& \begin{bmatrix} E(x_i^2) & 0 \\ 0 & E(w_i^2) \end{bmatrix} \\
E(v_i^2Z_iZ_i') &=& E(v_i^2 \begin{bmatrix} x_i \\ w_i \end{bmatrix} \begin{bmatrix} x_i & w_i \end{bmatrix}) \\
&=& \begin{bmatrix} E(v_i^2x_i^2) & E(v_i^2x_iw_i) \\ E(v_i^2x_iw_i) & E(v_i^2w_i^2) \end{bmatrix} \\
&=& \begin{bmatrix} E(v_i^2)E(x_i^2) & 0 \\ 0 & E(v_i^2w_i^2) \end{bmatrix} \\
&=& \begin{bmatrix} E(x_i^2)E(\epsilon_i^2)(1 - \rho_{w_i,\epsilon_i}^2) & 0 \\ 0 & E(v_i^2w_i^2) \end{bmatrix}
\end{align*}

If we focus only on $\hat \beta_1$, we have:

\begin{align*}
 \sqrt{n}(\hat \beta_l - \beta_1) \xrightarrow{d} N(0, E(\epsilon_i^2)(1 - \rho_{w_i,\epsilon_i}^2)/E(x_i^2))
\end{align*}

If we compare this variance to the short equation estimator, the comparison depends on the variance of $\beta w_i + \epsilon_i$, the variance of $\epsilon_i$ and the correlation coefficient $\rho_{w_i,\epsilon_i}$.

$$\frac{Var(\hat \beta_s)}{Var(\hat \beta_l)} = \frac{E((\epsilon_i+ \beta_2 w_i)^2)}{E(\epsilon_i^2)(1 - \rho_{w_i,\epsilon_i}^2)}$$

We can reexpress $\epsilon_i = \beta_3w_i + v_i$ and we will get the sum of two independent variables in the numerator:
\begin{align*}
\frac{Var(\hat \beta_s)}{Var(\hat \beta_l)} &=& \frac{E((v_i+ (\beta_2 +\beta_3) w_i)^2)}{E(\epsilon_i^2)(1 - \rho_{w_i,\epsilon_i}^2)} \\
&=& \frac{E(v_i^2)+ (\beta_2 +\beta_3)^2 E(w_i^2)}{E(\epsilon_i^2)(1 - \rho_{w_i,\epsilon_i}^2)} \\
&=& \frac{E(\epsilon_i^2)(1 - \rho_{w_i,\epsilon_i}^2)+ (\beta_2 +\beta_3)^2 E(w_i^2)}{E(\epsilon_i^2)(1 - \rho_{w_i,\epsilon_i}^2)} \ge 1\\
\end{align*}

Thus, the short equation estimator has a larger variance than the long equation estimator. The extreme case in which $\beta_2 = - \beta_3$ is related to equal variances.

### 2.3

In this setup:
\begin{align*}
\epsilon_i \perp x_i \to E(\epsilon_i \ x_i) = 0 \\
E(\epsilon_i w_i) \ne 0 \\
E(x_i w_i) \ne 0 \\
\end{align*}

We can make the computation of the cross expectations using these assumptions:
\begin{align*}
E(x_iy_i) &=& E(x_i (\beta_1 x_i + \beta_2 w_i + \epsilon_i)) \\
 &=& \beta_1E(x_i^2) + \beta E(x_i w_i)\\
E(w_iy_i) &=& E(w_i (\beta_1 x_i + \beta_2 w_i + \epsilon_i)) \\
 &=& \beta_1 E(w_i x_i) + \beta_2 E(w_i^2) + E(w_i \epsilon_i)\\
\end{align*}


##### (a)

Using the previous results, it is easy to show that (this case is equivalent to 2.1 ins spirit):

\begin{align*}
\hat\beta_s &\xrightarrow[p]{}& \frac{E(x_iy_i)}{E(x_i^2)} \\
\hat\beta_s &\xrightarrow[p]{}& \frac{\beta_1E(x_i^2) + \beta_2 E(x_i w_i)}{E(x_i^2)} \\
\hat\beta_s &\xrightarrow[p]{}& \beta_1 + \beta_2 \frac{E(x_i w_i)}{E(x_i^2)} \\
\end{align*}

Thus, the short regression is inconsistent.

##### (b)

Using the previous results

\begin{align*}
\hat\beta_l &\xrightarrow[p]{}& \frac{E(w_i^2)E(x_i y_i) - E(x_i w_i)E(w_i y_i)}{E(x_i^2)E(w_i^2) - E(x_iw_i)^2} \\
\hat\beta_l &\xrightarrow[p]{}& \frac{E(w_i^2)(\beta_1E(x_i^2) + \beta_2 E(x_i w_i)) - E(x_i w_i)E(\beta_1E(x_i w_i) + \beta_2 E(w_i^2) + E(w_i \epsilon_i))}{E(x_i^2)E(w_i^2) - E(x_iw_i)^2} \\
\hat\beta_l &\xrightarrow[p]{}& \frac{\beta_1(E(w_i^2)E(x_i^2) - E(x_i w_i)^2) - E(x_i w_i)E(w_i \epsilon_i)}{E(x_i^2)E(w_i^2) - E(x_iw_i)^2} \\
\hat\beta_l &\xrightarrow[p]{}& \beta_1 - \frac{E(x_i w_i)E(w_i \epsilon_i)}{E(x_i^2)E(w_i^2) - E(x_iw_i)^2}
\end{align*}
Thus, the long regression is inconsistent.

##### (c)

It is irrelevant as both estimators are inconsistent.

The unbiased value for this estimator is:

\begin{align*}
Var(\hat \beta) &=& E(X'X)^{-1} \sigma^2 \\
&=& \frac{1}{n(E(x_i^2) - E(x_i)^2)} \begin{bmatrix}
E(x_i^2) & -E(x_i) \\
-E(x_i) & 1
\end{bmatrix} \sigma^2 \\
&=&  \begin{bmatrix}
1/n & 0 \\
0 & 1/n
\end{bmatrix}
\end{align*}

```python
def get_standard_errors(i, sample = 1000):

    df = pd.DataFrame(index = range(0,sample))
    df['const'] = 1
    
    df['x'] = np.random.normal(0,1,sample)
    df['epsilon'] = np.random.normal(0,1,sample)
    df['y'] = df['x'] + df['const'] + df['epsilon']
    
    
    Y = df[['y']].to_numpy()
    X = df[['const','x']].to_numpy()
    
    beta = np.matmul(
    np.linalg.inv(np.matmul(np.transpose(X), X)),
    np.matmul(np.transpose(X),Y)
    )
    
    n, k = X.shape
    
    P = X @ np.linalg.inv(np.transpose(X) @ X) @ np.transpose(X)
    
    M = np.eye(n) - P
    
    e = M @ Y
    
    var_e = (np.transpose(e) @ e)/(n-k)
    
    var_b = var_e * np.linalg.inv(np.transpose(X) @ X)
    
    var_b_robust = (np.linalg.inv(np.transpose(X) @ X)
    @ np.transpose(X) @ (np.diag(e @ np.transpose(e) ) * np.eye(n)) @ X
    @ np.linalg.inv(np.transpose(X) @ X))


    var_est = {}

    
    var_est['c_nr'] = var_b[0,0]
    var_est['b_nr'] = var_b[1,1]
    var_est['c_r'] = var_b_robust[0,0]
    var_est['b_r'] = var_b_robust[1,1]
    
    return var_est
```

```python
def simulation_se(sample):

    #Simulation
    df_var_est = pd.DataFrame([*map(lambda y: get_standard_errors(y, sample), range(1000))])

    # Constant
    fig, ax = plt.subplots()
    sns.kdeplot(data=df_var_est['c_nr'],
    color='crimson', label='Non-robust', fill=True, ax=ax)
    sns.kdeplot(data=df_var_est['c_r'],
    color='limegreen', label='Robust', fill=True, ax=ax)
    ax.legend()
    ax.axvline(x = df_var_est['c_nr'].mean(), color='crimson', linestyle = '--', # Line on x = 2
    )
    ax.axvline(x = df_var_est['c_r'].mean(), color='limegreen', linestyle = '--', # Line on x = 2
    )
    ax.axvline(x = 1/sample, color='black', linestyle = '--', # Line on x = 2
    )
    plt.tight_layout()
    plt.show()

    # Beta coefficient
    
    fig, ax = plt.subplots()
    sns.kdeplot(data=df_var_est['b_nr'],
    color='crimson', label='Non-robust', fill=True, ax=ax)
    sns.kdeplot(data=df_var_est['b_r'],
    color='limegreen', label='Robust', fill=True, ax=ax)
    ax.legend()
    ax.axvline(x = df_var_est['b_nr'].mean(), color='crimson', linestyle = '--', # Line on x = 2
    )
    ax.axvline(x = df_var_est['b_r'].mean(), color='limegreen', linestyle = '--', # Line on x = 2
    )
    ax.axvline(x = 1/sample, color='black', linestyle = '--', # Line on x = 2
    )
    plt.tight_layout()
    plt.show()
```

```python
simulation_se(30)
```

```python
simulation_se(100)
```

```python
simulation_se(1000)
```

```python
simulation_se(10000)
```

We can observe that, in small samples, the non-robust standard errors have more bias than the robust standard errors. If the sample is small, the size of the bias is larger. When we increase the sample size, this bias decreases. 

Asymptotically, both non-robust and robust errors are unbiased. So, in that case, we prefer the non-robust standard errors because they have less variance.
