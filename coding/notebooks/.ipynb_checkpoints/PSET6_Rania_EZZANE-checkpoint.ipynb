#### 5. Question
The code simulates an ARMA(1,1) process using the given parameters $\phi=0.9$ (autoregressive coefficient), $\gamma_1=0.1$ (moving average coefficient), $\sigma^2=1$ (variance of the white noise), and $T=150$ (number of time points). It generates the ARMA(1,1) process iteratively and then plots the simulated ARMA(1,1) process.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import norm as norm

def calculate_log_likelihood_ARMA1(data, variance_e, psi):
    T = len(data)
    sum_of_squares = 0
    for t in range(1, T):
        error_term = data[t] - psi * data[t-1]
        sum_of_squares += error_term ** 2

    sum_of_terms = sum_of_squares / (2 * variance_e) + (T - 1) / 2 * np.log(2 * np.pi * variance_e)
    return -sum_of_terms

# Set seed for reproducibility
np.random.seed(2024)

# Define parameters
psi = 0.9
gamma1 = 0.1
sigma2 = 1
T = 150

# Generate white noise
epsilon = np.random.normal(0, sigma2, T)
epsilon_lag1 = np.roll(epsilon, 1)
epsilon_lag1[0] = 0

# Generate ARMA(1,1) process
y_data = np.zeros(T)
for t in range(1, T):
    y_data[t] = psi * y_data[t-1] + epsilon[t] + gamma1 * epsilon_lag1[t]

# Plot simulated ARMA(1,1) process
plt.plot(y_data)
plt.title('Simulated ARMA(1,1) Process')
plt.xlabel('Time, $t$')
plt.ylabel('Dependent variable, $y_{t}$')
plt.show()
```

It iterates through each time step, updating the predicted state and covariance matrices using Kalman filter equations. It also calculates the likelihood of each observation at each time step based on the Kalman filter estimates. The function returns the log likelihood, predicted state, and covariance matrices.

The code computes the log likelihood of the simulated data using both the Kalman filter and a theoretical log-likelihood function $\texttt{theoretical$\_$log$\_$likelihood$\_$ARMA1()}$. The theoretical log likelihood function calculates the log likelihood based on the ARMA(1,1) model parameters.

The numerical routine implemented in the Kalman filter takes the model parameters $\theta$ and the series $\mathbf{y}_T$ as inputs and returns the log likelihood function $L_t$, the estimated state vector $\mathbf{Z}_t$, and the estimated state covariance matrices $\mathbf{P}_t$.

Now, let's discuss how it compares to the likelihood function computed in point 3:
 - The likelihood function computed in point 3 is calculated using the theoretical_log_likelihood_ARMA1() function, which provides a theoretical log-likelihood based on the ARMA(1,1) model parameters.
- The likelihood function computed by the Kalman filter is based on the observed data and the estimated states obtained iteratively using the Kalman filter algorithm.


```python
def kalman_filter(y_data, theta):
    # Unpack parameters
    phi, gamma1, sigma2 = theta

    T = len(y_data)

    # Define matrices
    A = np.array([[phi, 1, gamma1]])
    H = np.zeros((1, 3))
    Phi = np.array([[phi, 1, gamma1], [0, 0, 0], [0, 1, 0]])
    R = np.array([[0, 0, 0], [1, 0, 0], [0, 0, 0]])
    Sigma = np.eye(3) * sigma2

    # Initialize arrays
    X_t = np.zeros((T, 3))
    V_t = np.zeros((T, 3, 3))
    L_t = np.zeros(T)

    # Initial state estimation
    x0 = np.zeros(3)
    V0 = np.array([[sigma2 * (1 + 2 * gamma1 * phi + gamma1 ** 2) / (1 - phi ** 2), 0, sigma2],
                   [0, sigma2, 0],
                   [sigma2, 0, sigma2]])
    pred_X_t_given_X_t1 = np.zeros((T, 3))
    pred_X_t_given_X_t1[0] = Phi @ x0

    # Kalman filter iteration
    for t in range(T):
        # Predicted state and prediction error
        pred_X = pred_X_t_given_X_t1[t]
        y_hat = np.dot(A, pred_X)
        v_t = y_data[t] - y_hat

        # Calculate Kalman gain and update state estimation
        pred_V = Phi @ V_t[t - 1] @ Phi.T + R @ Sigma @ R.T if t > 0 else V0
        F_t = np.dot(np.dot(A, pred_V), A.T) + np.dot(np.dot(H, H.T), sigma2)
        K_t = np.dot(pred_V @ A.T, np.linalg.inv(F_t))
        X_t[t] = pred_X + np.dot(K_t, v_t)

        # Calculate updated state covariance
        V_t[t] = pred_V - np.dot(np.dot(K_t, A), pred_V)

        # Update prediction for the next time step
        if t < T - 1:
            pred_X_t_given_X_t1[t + 1] = Phi @ X_t[t]

        # Compute log-likelihood
        L_t[t] = norm.logpdf(y_data[t], loc=y_hat, scale=np.sqrt(F_t))

    return L_t, V_t, X_t

# Define parameters
phi = 0.9
gamma1 = 0.1
sigma2 = 1
theta = [phi, gamma1, sigma2]

# Apply the Kalman filter
L_t, V_t, X_t = kalman_filter(y_data, theta)

print("Log-likelihood Theoreticql:", calculate_log_likelihood_ARMA1(y_data, sigma2, phi))
print("Log-likelihood from Kalman Filter:", np.sum(L_t))
```

```python
# Set seed for reproducibility
np.random.seed(2024)

# Define parameters
phi_value = 0.9
gamma_value = 0.1
sigma_squared = 1
time_steps = 150

# Generate white noise
noise = np.random.normal(0, sigma_squared, time_steps)
lagged_noise = np.roll(noise, 1)
lagged_noise[0] = 0

# Generate ARMA(1,1) process
y_series = np.zeros(time_steps)
for t_index in range(1, time_steps):
    y_series[t_index] = phi_value * y_series[t_index-1] + noise[t_index] + gamma_value * lagged_noise[t_index]


def kalman_filter(data, parameters):
    # Unpack parameters
    phi, gamma, sigma_sq = parameters

    T = len(data)

    # Define matrices
    A = np.array([[phi, 1, gamma]])
    H = np.zeros((1, 3))
    Phi = np.array([[phi, 1, gamma], [0, 0, 0], [0, 1, 0]])
    R = np.array([[0, 0, 0], [1, 0, 0], [0, 0, 0]])
    Sigma = np.eye(3) * sigma_sq

    # Initialize arrays
    X_t = np.zeros((T, 3))
    V_t = np.zeros((T, 3, 3))
    L_t = np.zeros(T)

    # Initial state estimation
    initial_state = np.zeros(3)
    initial_variance = np.array([[sigma_sq * (1 + 2 * gamma * phi + gamma ** 2) / (1 - phi ** 2), 0, sigma_sq],
                                 [0, sigma_sq, 0],
                                 [sigma_sq, 0, sigma_sq]])
    pred_state_given_state_prev = np.zeros((T, 3))
    pred_state_given_state_prev[0] = Phi @ initial_state

    # Kalman filter iteration
    for t in range(T):
        # Predicted state and prediction error
        pred_state = pred_state_given_state_prev[t]
        y_predicted = np.dot(A, pred_state)
        prediction_error = data[t] - y_predicted

        # Calculate Kalman gain and update state estimation
        pred_variance = Phi @ V_t[t - 1] @ Phi.T + R @ Sigma @ R.T if t > 0 else initial_variance
        innovation_covariance = np.dot(np.dot(A, pred_variance), A.T) + np.dot(np.dot(H, H.T), sigma_sq)
        kalman_gain = np.dot(pred_variance @ A.T, np.linalg.inv(innovation_covariance))
        X_t[t] = pred_state + np.dot(kalman_gain, prediction_error)

        # Calculate updated state covariance
        V_t[t] = pred_variance - np.dot(np.dot(kalman_gain, A), pred_variance)

        # Update prediction for the next time step
        if t < T - 1:
            pred_state_given_state_prev[t + 1] = Phi @ X_t[t]

        # Compute log-likelihood
        L_t[t] = norm.logpdf(data[t], loc=y_predicted, scale=np.sqrt(innovation_covariance))

    return L_t, V_t, X_t

# Define parameters for the filter
phi_val = 0.9
gamma_val = 0.1
sigma_squared_val = 1
theta_params = [phi_val, gamma_val, sigma_squared_val]

# Apply the Kalman filter
log_likelihoods, state_variance, state_estimations = kalman_filter(y_series, theta_params)

# Plotting the log likelihood as a function of phi
n_points = 100
phi_values = np.linspace(-1, 1, n_points)
likelihood_values = np.zeros(n_points)

for index, phi_val in enumerate(phi_values):
    theta_params = np.array([phi_val, gamma_value, sigma_squared])
    log_likelihoods, _, _ = kalman_filter(y_series, theta_params)
    sum_log_likelihood = log_likelihoods.sum()
    likelihood_values[index] = sum_log_likelihood
    if phi_val == phi_values[0]:
        max_likelihood = sum_log_likelihood
        max_phi = phi_val
    elif sum_log_likelihood > max_likelihood:
        max_likelihood = sum_log_likelihood
        max_phi = phi_val

plt.plot(phi_values, likelihood_values)
plt.title('Log likelihood as a function of $\phi$')
plt.xlabel('$\phi$')
plt.ylabel('Log Likelihood')
plt.show()

from scipy.optimize import minimize

def log_likelihood_func(theta, data):
    log_likelihoods, _, _ = kalman_filter(data, theta)
    return log_likelihoods.sum()

def log_prior_func(theta):
    phi_val = theta[0]
    gamma_val = theta[1]
    sigma_squared_val = theta[2]
    if np.abs(phi_val) < 1 and sigma_squared_val > 0:
        return 0
    else:
        return -1e10

def log_posterior_func(theta, data):
    return log_likelihood_func(theta, data) + log_prior_func(theta)

# Minimize the negative of posterior
def neg_posterior_func(theta, data):
    return -1 * log_posterior_func(theta, data)

# Initial guess
initial_theta = np.array([0.9, 0.1, 1])

# Bounds
parameter_bounds = [(-0.99, 0.99), (-0.99, 0.99), (0.01, None)]

# Minimizer
result = minimize(neg_posterior_func, initial_theta, args=(y_series), bounds=parameter_bounds, method='L-BFGS-B')
optimal_theta = result.x
print('Optimal Theta:\n', pd.DataFrame(result.x.reshape(1, 3), columns=['phi', 'gamma', 'sigma_squared']))

negative_hessian = result.hess_inv.todense()
inverted_negative_hessian = np.linalg.inv(negative_hessian)

print('Negative Hessian:\n', pd.DataFrame(negative_hessian))

```

#### 2. 
Let us write a computer program that generates draws from the Metropolis-Hastings al- gorithm that we have seen in class. For that purpose, we use the candidate density

$\begin{equation}
q(\theta^{n1})  \mathcal{N}(\theta^{n1}, c\Sigma)
\end{equation}$,

where c is a constant, and choose c so that you accept roughly $30\%$ of the draws.

```python
def metropolis_hastings(initial_theta, iterations, c, Sigma, target_acceptance_rate, data):
    np.random.seed(2024)
    current_theta = initial_theta
    theta_samples = np.zeros((iterations, 3))
    acceptance_count = 0

    for iteration in range(iterations):
        if iteration % (iterations // 20) == 0:
            print('Progress:', iteration / iterations)
        
        proposed_theta = np.random.multivariate_normal(current_theta, c * Sigma)
        alpha = np.exp(log_posterior_func(proposed_theta, data) - log_posterior_func(current_theta, data))
        
        if np.random.uniform(0, 1) < alpha:
            current_theta = proposed_theta
            acceptance_count += 1
        else:
            acceptance_count += 0
        
        theta_samples[iteration] = current_theta
        
        if iteration % 100 == 0 and iteration > 0:
            current_acceptance_rate = acceptance_count / iterations
            c = c * (current_acceptance_rate / target_acceptance_rate)
    
    acceptance_rate = acceptance_count / iterations
    print('Acceptance rate:', acceptance_rate)
    
    return theta_samples, acceptance_rate
```

The provided Python function implements the Metropolis-Hastings algorithm, a Markov Chain Monte Carlo (MCMC) method, to generate samples from the posterior distribution of a Bayesian model. The function takes several arguments: the initial parameter values $\texttt{initial_theta}$, the number of samples to generate $\texttt{num_samples}$, the scale factor for the proposal distribution $\texttt{scale_factor}$, the covariance matrix for the proposal distribution $\texttt{covariance_matrix}$, the target acceptance rate $\texttt{target_acceptance_rate}$, and the observed data $\texttt{observed_data}$. 

During each iteration of the algorithm, a proposed parameter vector is generated by sampling from a multivariate normal distribution centered at the current parameter vector with covariance determined by $\texttt{covariance\_matrix}$ scaled by $\texttt{scale_factor}$. The acceptance probability of the proposed parameter vector is computed based on the ratio of the posterior densities of the proposed and current parameter vectors. If the acceptance probability exceeds a random threshold, the proposed parameter vector is accepted as the new current parameter vector; otherwise, the current parameter vector remains unchanged. 

The algorithm adjusts the $\texttt{scale_factor}$ dynamically to achieve a desired acceptance rate close to $\texttt{target_acceptance_rate}$. The generated parameter samples and the acceptance rate are returned as output.
